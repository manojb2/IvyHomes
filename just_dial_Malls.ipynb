```python
def strings_to_num(argument): 
    
    switcher = { 
        'dc': '+',
        'fe': '(',
        'hg': ')',
        'ba': '-',
        'acb': '0', 
        'yz': '1', 
        'wx': '2',
        'vu': '3',
        'ts': '4',
        'rq': '5',
        'po': '6',
        'nm': '7',
        'lk': '8',
        'ji': '9'
    } 
    
    return switcher.get(argument, "nothing")
```

```python
link = "https://www.justdial.com/Bangalore/Malls/nct-10310530/page-"
```

```python
 import re
```

```python
import requests
 
```

```python
nameList = []
addressList = []
numbersList = []
votesList =[]
ratingsList = []
latitudeList = []
longitudeList = []
    #tempList=[]
```

```python
p =1
while p<11:
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get(link+ str(p))
    p=p+1
    driver.implicitly_wait(10)
    from selenium.webdriver.remote.webelement import WebElement
    SCROLL_PAUSE_TIME = 3

    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")
    i=1
    y = 2000
    while i<13:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight-900);")
        i=i+1
        y=y+2000
        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)
        
        
        
    storeDetails = driver.find_elements_by_class_name('store-details')


   

    for i in range(len(storeDetails)):

        name = storeDetails[i].find_element_by_class_name('lng_cont_name').text
        address = storeDetails[i].find_element_by_class_name('cont_fl_addr').get_attribute('innerHTML')
        contactList = storeDetails[i].find_elements_by_class_name('mobilesv')
        temp = storeDetails[i].find_elements_by_class_name('rating_div')[1].get_attribute('text')
        temp=temp.replace("\t","")
        temp=temp.replace("\n","")
        elem = storeDetails[i].find_element_by_class_name('rating_div').get_attribute('href')
        ratings = temp[:3]
        votes = temp[3:]
        #votes = votes.replace("\n","")
        #ratings = storeDetails[i].find_elements_by_class_name('rating_div')[1].get_attribute('text')
        s = requests.Session()

        url = "https://www.justdial.com/functions/maps.php"

        hdr = {
                'accept': 'application/json, text/javascript, */*; q=0.01',
                'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36',
                'content-type': 'application/x-www-form-urlencoded; charset=UTF-8',
                'origin': 'https://www.justdial.com',
                'referer': elem,
                'accept-encoding': 'gzip, deflate, br',
                'accept-language': 'en-US,en;q=0.9'
                }

        
        s1 = re.search(r'/080(.*?)_BZDET', elem).group(1)
        s1 = s1.replace("-",".")
        s1 = "080" + s1;
        formdata = {
                'doc': s1
                }

        resp = s.post(url, headers=hdr, data=formdata)
        loc = resp.text
        latitude = re.search(r'l":"(.*?)","',loc).group(1)
        longitude = re.search(r'n":"(.*?)"}',loc).group(1)
        

        myList = []

        for j in range(len(contactList)):

            myString = contactList[j].get_attribute('class').split("-")[1]

            myList.append(strings_to_num(myString))

        nameList.append(name)
        addressList.append(address)
        numbersList.append("".join(myList))
        votesList.append(votes)
        ratingsList.append(ratings)
        latitudeList.append(latitude)
        longitudeList.append(longitude)
```

```python
data = {'Company Name':nameList,
        #'temp':tempList,
        'Address': addressList,
        'Phone':numbersList,
        'votes':votesList,
        'ratings':ratingsList,
        'latitude':latitudeList,
        'longitude':longitudeList
        
       }

# Create DataFrame
df = pd.DataFrame(data)
```
